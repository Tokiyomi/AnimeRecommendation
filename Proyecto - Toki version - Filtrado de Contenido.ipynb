```python
#Calcular las puntuaciones de este usuario hipotetico (se podria usar np.dot pero lo hice a mano)
def dot_product(vector_1, vector_2):
    """Funcion que realiza el producto punto entre dos matrices
        ARGS: 
            vector_1: una matriz de nxm
            vector_2: una matriz de nxm
        RETURNS:
            La suma del producto punto de todos los elementos"""
    return sum([ i*j for i,j in zip(vector_1, vector_2)])

def get_anime_score(anime_features, user_preferences):
    """Funcion que genera el score de un título basadoe en las preferencias del usuario
        ARGS:
            anime_features: una matriz de 1's y 0's dependiendo de las caracteristicas propias del título
            user_preferences: una matriz de enteros del 1 al 10 deprendiendo de las preferencias del usuario por género
        RETURNS:
            El producto punto de estas dos matrices que representan el score final"""
    return (dot_product(anime_features, user_preferences))
```
